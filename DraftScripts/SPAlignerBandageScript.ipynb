```python
#Create a dataframe for SPAligner Combined1
SPAligner_Combined1=pd.read_csv('SPAligneroutputcombined1.tsv', sep='\t')
```

## This section cleans the data from SPAligner so that the columns can be compared easily.

```python
#format the Query column from  SPAligner_Combined1 so that the Query column is the same as the one from Bandage
pattern = r'\[.*?\]'
def format_query(x):
    return re.sub(pattern,"", x).rstrip()
    


SPAligner_Combined1['Query'] = SPAligner_Combined1['Query'].map(format_query)
```

## This section creates 2 dataframes so that it is easier to compare. The Query and Resulting path are kept.

```python
#Get the query and sequence column from Bandage_Combined1
Bandage_Combined_Query_Path =Bandage_Combined1[['Query','Sequence']]
Bandage_Combined_Query_Path.head()

```

```python
#Get the query and sequence column from SPAligner_Combined1
SPAligner_Combined_Query_Path=SPAligner_Combined1[['Query','Sequence']]
SPAligner_Combined_Query_Path.head()
```

## This section creates df_union. The dataframe stores the Query and path from both software.

```python
#Use merge operation so that we have the sequences for both softwares for successful queries
#df=pd.merge(Bandage_Combined_Query_Path,SPAligner_Combined_Query_Path, on='Query')
df_union=pd.merge(Bandage_Combined_Query_Path,SPAligner_Combined_Query_Path, on='Query',how="outer")
#Paths, start position etc
```

```python
df_union.rename(columns = {'Sequence_x':'Path_Bandage'}, inplace = True)
df_union.rename(columns = {'Sequence_y':'Path_SPAligner'}, inplace = True)
```

```python
df_union
```

```python
df_union.to_csv("test2.csv")
```

```python
#Check for similar path for the queries
df_union['Equal'] = np.where(df_union['Path_Bandage']==df_union['Path_SPAligner'], 
                                           'yes', 'no')
```

## This section checks the queries that do not have any path in SPAligner.

```python
#Check for similar path for the queries
df_union['Query_Not_SPAligner'] = np.where(df_union['Path_SPAligner'].isna(), 
                                           'yes', 'no')
```
