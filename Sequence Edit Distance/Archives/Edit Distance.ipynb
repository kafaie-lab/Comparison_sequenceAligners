```python
import pandas as pd
import Levenshtein
from Bio import SeqIO
```

```python
#Read the query and sequences from the .fasta file
fasta_file = "REAL/ERR.fasta"  

# Create empty lists to store the data
queries = []
sequences = []

# Read through the FASTA file and extract query and sequence information
for record in SeqIO.parse(fasta_file, "fasta"):
    queries.append(record.id)
    sequences.append(str(record.seq))

# Create DataFrame from the lists
fasta_df = pd.DataFrame({'Query': queries, 'Sequence': sequences})

# Display the DataFrame
print(fasta_df)
```

```python
fasta_df=fasta_df.sort_values("Query")
fasta_df=fasta_df.reset_index(drop=True)
fasta_df
```

```python
fasta_df.to_csv("testSequences.tsv",sep="\t")
```

```python
#Discard the colums Sequence Bandage
df1_bandage=df1_bandage.drop(['Sequence_Bandage'], axis=1)
```

```python
merged_df = pd.merge(fasta_df, df1_bandage,on='Query')
```

```python
merged_df
```

```python
merged_df.rename(columns = {'Sequence_x':'Sequence'}, inplace = True)
merged_df.rename(columns = {'Sequence_y':'Sequence_Bandage'}, inplace = True)
```

```python
merged_df.dtypes
```

```python
merged_df['Sequence_y']=merged_df['Sequence_Bandage'].astype(str)
```

```python
#Create a method to calculate the levenstein distance
def calculate_edit_distance_query_bandage(row):
    return Levenshtein.distance(row['Sequence'], row['Sequence_Bandage'])
```

```python
merged_df['edit_distance_query_bandage'] = merged_df.apply(calculate_edit_distance_query_bandage, axis=1)
```

```python
merged_df
```

```python
#Create a method to calculate the levenstein distance
def calculate_edit_distance_query_spaligner(row):
    return Levenshtein.distance(row['Sequence'], row['Sequence_SPAligner'])
```

```python
merged_df['edit_distance_query_spaligner'] = merged_df.apply(calculate_edit_distance_query_spaligner, axis=1)
```

```python
merged_df
```

```python
#Create a method to calculate the levenstein distance
def calculate_edit_distance_bandage_spaligner(row):
    return Levenshtein.distance(row['Sequence_Bandage'], row['Sequence_SPAligner'])
```

```python
merged_df['edit_distance_query_bandage_spaligner'] = merged_df.apply(calculate_edit_distance_bandage_spaligner, axis=1)
```
