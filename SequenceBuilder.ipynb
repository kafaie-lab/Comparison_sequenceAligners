```python
import gfapy
import re
from gfapy.sequence import rc

gfaFile = input("Enter the address of the gfa file: ")
#Loading the graph
try:
    print('Loading the graph ...')
    myGraph = gfapy.Gfa.from_file(gfaFile)
except Exception as e:
    print("Graph not loaded successfully: " + str(e))
    import pdb
    pdb.set_trace()

path = input("Enter the node_list, start_pos and end_pos seperated by space: ")
while path!='-1' and path != 'exit':
    list = path.split()
    nodeOrients = list[0]
    start = int(list[1])
    end = int(list[2])
    orientList = []
    for ch in nodeOrients:
        if ch=='>':
            orientList.append('+')
        elif ch == '<':
            orientList.append('-')
    nodeList = re.split('>|<', nodeOrients)
    nodeList = nodeList[1:]

    print('Generating the sequence ...')
    sequence =''
    counter = 1
    for node, orient in zip(nodeList, orientList):
        seq = myGraph.segment(node).sequence
        if orient == '-':
            seq =  rc(seq)
        if counter == 1:
            sequence = seq[start:]
        elif counter == len(nodeList):
            sequence += seq[55:end]
        else:
            sequence +=seq[55:]

        counter=counter+1

    print(sequence)
    path = input("Enter the node_list, start_pos and end_pos seperated by space: ")
```
